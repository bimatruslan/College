### Langkah 4 Konversi data
Dikarenakan model prediksi tidak dapat memprose data dengan _value string_, maka perlu dikonversi variable yang memiliki nilai _string_ menjadi numerikal.

Dalam kasus ini terdapat dua jenis konversi yang akan dilakukan yaitu :
    
    - Perubahan value kategorik yang memiliki 2 tipe (_binary_) menjadi 0 dan 1. Variable yang dirubah :
        - Married             = (Yes, No) Yes menjadi 1 dan No menjadi 0
        - Education           = (Graduate, Not Graduate) Graduate menjadi 1 dan Not Graduate menjadi 0
        - Self_Employed       = (Yes, No) Yes menjadi 1 dan No menjadi 0
        - Loan_Status         = (Y,N) Y menjadi 1 dan N menjadi 0
    
    - Perubahan value kategorik yang memiliki jenis lebih dari 2 atau bukan yes no, maka akan menggunakan dummy. Variable yang dirubah :
        - Gender              = (Male, Female)
        - Dependents          = (0â€“3)
        - Property_Area       = (Urban: Perkotaan, Semiurban: Semi perkotaan, Rural: Pedesaan)
        
    - Merubah Credit_History dari Float menjadi tipe int

```python
data_copy['Self_Employed'] = data_copy['Self_Employed'].replace({'No': 0, 'Yes': 1})
data_copy['Married'] = data_copy['Married'].replace({'No': 0, 'Yes': 1})
data_copy['Education'] = data_copy['Education'].replace({'Not Graduate': 0, 'Graduate': 1})
data_copy['Loan_Status'] = data_copy['Loan_Status'].replace({'N': 0, 'Y': 1})
data_copy
```

```python
data_copy = data_copy.join(pd.get_dummies(data_copy['Gender']))
data_copy = data_copy.join(pd.get_dummies(data_copy['Dependents']))
data_copy = data_copy.join(pd.get_dummies(data_copy['Property_Area']))
data_copy = data_copy.drop(columns = 'Gender')
data_copy = data_copy.drop(columns = 'Dependents')
data_copy = data_copy.drop(columns = 'Property_Area')
data_copy
```

```python
data_copy['Credit_History'] = data_copy['Credit_History'].astype('int64')
data_copy.info()
```

## Langkah Pemabgian Data Training dan Testing
Pada langkah ini akan menggunakan _library_ __sklearn__ maka langkah pertama yang dilakukan adalah mengimport _library_ tersebut

```python
from sklearn import model_selection #Library untuk proses prediksi
from sklearn.model_selection import KFold # Cross Validation
from sklearn.model_selection import cross_val_score # Mengambil nilai hasil tes
from sklearn.model_selection import cross_validate # Mengambil nilai hasil tes
from sklearn.model_selection import train_test_split # Repeated Hold Out
from sklearn.metrics import classification_report # Menampilkan hasil tes
```

## Pembagian Data Training dan Testing
Dalam pembagian dataset ini perlu ditentukan independent variable (y) dan dependent variable(x) yang akan digunakan. Dalam hal ini seluruh variable selain Loan ID dan Loan Status. Loan Status akan digunakan sebagai dependent variable(x)

```python
fitur = ["Married", "Education", "Self_Employed", "ApplicantIncome", "CoapplicantIncome", "LoanAmount", "Loan_Amount_Term", "Credit_History", 
         "Female", "Male", "0", "1", "2", "3+", "Rural", "Semiurban", "Urban"]

x = data_copy[fitur]
y = data_copy['Loan_Status']
            
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, shuffle=True, random_state = 24)
```

## Membangun Model Prediksi
Dalam kasus ini metode yang akan digunakan dalam model adalah Naive Bayes. Kfold akan digunakan untuk membuat iterasi _testing_ yang akan dilakukan untuk mendapatkan nilai kinerja dari model.

```python
from sklearn.naive_bayes import GaussianNB
naivebayes = GaussianNB()
kfold = model_selection.KFold()
```

## Melatih Model
Pada langkah ini model prediksi yang sudah dibuat akan di latih menggunakan data training yang sudah dibuat pada langkah sebelumnya. Training dilakukan dengan meggunakan _function_ .__fit()__

```python
naivebayes.fit(x_train, y_train)
```

## Evaluasi Kinerja Model
Dalam tahap ini metriks yang digunakan untuk mengukur kinerja dari model tersebut adalah _Accuracy_ dan _f1-measure_ . Untuk mendapatkan nilai dari kedua metriks tersebut digunakan metode _Cross Validation_

```python
score_test = naivebayes.score(x_test, y_test)
predict = naivebayes.predict(x_test)
print(classification_report(y_test, predict))
```

```python
scoring = ['accuracy', 'f1']
print(cross_val_score(naivebayes, x, y, cv=kfold, scoring="accuracy"))

result = cross_val_score(naivebayes, x, y, cv=kfold, scoring="accuracy")
print(result.mean())
```

Nilai diatas merupakan hasi iterasi testing dari model yang sudah dibuat, dari hasil tersebut akan di rata-rata untuk mengetahui kinerja model secara menyeluruh sebesar 0.78 atau 78%

```python
train_size = [0.9, 0.8, 0.75, 0.7, 2/3]
for t in train_size :
    naivebayes = GaussianNB()
    x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=t, random_state=24)
    naivebayes.fit(x_train, y_train)
    score = naivebayes.score(x_test, y_test)
    print("Accuracy Naive Bayes with train size %.2f : %.3f%%" % (t, score*100))
```
